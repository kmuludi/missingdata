```python
def jarak(df2,col):
    euc_dist[col] = []
    for row in range(df2.shape[0]):
        current = df2.index[df2.iloc[:,col] == 0].tolist()
        current = current[0]
        a = df2.iloc[current,col+1:].tolist()
        b = df2.iloc[current,:col].tolist()
        x1 = b+a
        if(row is not current):
            a=df2.iloc[row,col+1:].tolist()
            b=df2.iloc[row,:col].tolist()
            x2=b+a
        else:
            continue
        dist = euclidean(x1,x2)
        euc_dist[col].append([dist,df2.iloc[row,col]])
        print(" x1 = ",x1,"\n x2 = ",x2,"\n distance = ",euc_dist[col])

    lst = list(euc_dist[col])
    lst.sort()
    euc_dist[col] = lst
    print(euc_dist[col])
```

```python
def gaussian(x, mean, std_dev):
    return np.exp(-1 * (x-mean)**2 / (2 * std_dev**2))
```

# Klasifikasi

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
x = dummy.iloc[:, :-1].values
y = dummy.iloc[:, -1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
