```python
import matplotlib.pyplot as plt

# Persentase akurasi
persentase = [10, 20, 30, 40]

# Data akurasi
mean = [93.82, 93.09, 92.73, 92.73]
median = [93.82, 93.09, 92.73, 92.73]
data_asli = [92.13, 92.59, 92.59, 92.59]
fuzzy_segitiga_k3 = []
fuzzy_segitiga_k5 = []
fuzzy_segitiga_k7 = []
fuzzy_segitiga_k9 = []
fuzzy_gaussian_k3 = []
fuzzy_gaussian_k5 = []
fuzzy_gaussian_k7 = []
fuzzy_gaussian_k9 = []

# Membuat grafik
plt.plot(persentase, mean, marker='o', label='Mean')
plt.plot(persentase, median, marker='o', label='Median')
plt.plot(persentase, data_asli, marker='o', label='Data Asli')
plt.plot(persentase, fuzzy_segitiga_k3, marker='o', label='Fuzzy Segitiga k=3')
plt.plot(persentase, fuzzy_segitiga_k5, marker='o', label='Fuzzy Segitiga k=5')
plt.plot(persentase, fuzzy_segitiga_k7, marker='o', label='Fuzzy Segitiga k=7')
plt.plot(persentase, fuzzy_segitiga_k9, marker='o', label='Fuzzy Segitiga k=9')
plt.plot(persentase, fuzzy_gaussian_k3, marker='o', label='Fuzzy Gaussian k=3')
plt.plot(persentase, fuzzy_gaussian_k5, marker='o', label='Fuzzy Gaussian k=5')
plt.plot(persentase, fuzzy_gaussian_k7, marker='o', label='Fuzzy Gaussian k=7')
plt.plot(persentase, fuzzy_gaussian_k9, marker='o', label='Fuzzy Gaussian k=9')

# Mengatur label sumbu x dan y
plt.xlabel('Klasifikasi KNN')
plt.ylabel('Persentase Akurasi')

# Mengatur judul grafik
plt.title('Hepatitis Dataset')

# Menambahkan legenda
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')

# Menambahkan label pada sumbu x
plt.xticks(persentase, ['k=3', 'k=5', 'k=7', 'k=9'])

# Menampilkan grafik
plt.show()
```
