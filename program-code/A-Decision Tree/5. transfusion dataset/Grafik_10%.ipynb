```python
import matplotlib.pyplot as plt

# Persentase akurasi
persentase = [10, 20, 30, 40]

# Data akurasi
mean = [71.66, 70.59, 71.66, 69.52]
median = [67.91, 70.59, 71.12,69.52]
data_asli = [64.86, 69.37, 65.77, 63.96]
fuzzy_segitiga_k3 = [84.49, 83.96, 80.75, 82.35]
fuzzy_segitiga_k5 = [79.14, 82.35, 82.35, 81.28]
fuzzy_segitiga_k7 = [80.21, 81.28, 79.68, 79.14]
fuzzy_segitiga_k9 = [79.14, 81.82, 81.28, 81.28]
fuzzy_gaussian_k3 = [82.89, 82.89, 81.82, 80.75]
fuzzy_gaussian_k5 = [78.07, 81.82, 81.82, 81.82]
fuzzy_gaussian_k7 = [79.14, 79.68, 80.75, 77.01]
fuzzy_gaussian_k9 = [79.14, 81.28, 79.14, 78.07]

# Membuat grafik
plt.plot(persentase, mean, marker='o', label='Mean')
plt.plot(persentase, median, marker='o', label='Median')
plt.plot(persentase, data_asli, marker='o', label='Listwise Deletion')
plt.plot(persentase, fuzzy_segitiga_k3, marker='o', label='Triangular MF k=3')
plt.plot(persentase, fuzzy_segitiga_k5, marker='o', label='Triangular MF k=5')
plt.plot(persentase, fuzzy_segitiga_k7, marker='o', label='Triangular MF k=7')
plt.plot(persentase, fuzzy_segitiga_k9, marker='o', label='Triangular MF k=9')
plt.plot(persentase, fuzzy_gaussian_k3, marker='o', label='Gaussian MF k=3')
plt.plot(persentase, fuzzy_gaussian_k5, marker='o', label='Gaussian MF k=5')
plt.plot(persentase, fuzzy_gaussian_k7, marker='o', label='Gaussian MF k=7')
plt.plot(persentase, fuzzy_gaussian_k9, marker='o', label='Gaussian MF k=9')

# Mengatur label sumbu x dan y
plt.xlabel('KNN Classification')
plt.ylabel('Percentage Accuracy')

# Mengatur judul grafik
plt.title('Blood Transfusion Dataset 10%')

# Menambahkan legenda
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')

# Menambahkan label pada sumbu x
plt.xticks(persentase, ['k=3', 'k=5', 'k=7', 'k=9'])

# Menampilkan grafik
plt.show()
```
