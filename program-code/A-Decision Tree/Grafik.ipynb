```python
import matplotlib.pyplot as plt

# Persentase akurasi
persentase = [10, 20]

# Data akurasi
mean = [74.50, 79.25]
imputasi_k3 = [96, 97]
imputasi_k5 = [94, 98]
imputasi_k7 = [94, 97]
imputasi_k9 = [95, 97]

# Membuat grafik
plt.plot(persentase, mean, marker='o', label='Mean')
plt.plot(persentase, imputasi_k3, marker='o', label='KNN k=3')
plt.plot(persentase, imputasi_k5, marker='o', label='KNN k=5')
plt.plot(persentase, imputasi_k7, marker='o', label='KNN k=7')
plt.plot(persentase, imputasi_k9, marker='o', label='KNN k=9')

# Mengatur label sumbu x dan y
plt.xlabel('Classification')
plt.ylabel('Percentage Accuracy')

# Mengatur judul grafik
plt.title('Chronic Kidney Dataset')

# Menambahkan legenda
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')

# Menambahkan label pada sumbu x
plt.xticks(persentase, ['Naive Bayes', 'Neural Network'])

# Menampilkan grafik
plt.show()
```
