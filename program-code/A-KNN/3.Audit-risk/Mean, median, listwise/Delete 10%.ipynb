```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("audit_risk.csv")
```

```python
df
```

```python
pct_missing = 0.104
num_missing = int(df.iloc[:, :-1].size * pct_missing)
num_missing
```

```python
df.iloc[:, :-1].size
```

```python
idx = np.random.choice(df.iloc[:, :-1].size, num_missing, replace=False)
```

```python
dataset = df.astype(float)
dataset.iloc[:, :-1].values.flat[idx] = np.nan
```

```python
dataset
```

```python
dataset.to_csv('audit_10%.csv', index=False)
```
